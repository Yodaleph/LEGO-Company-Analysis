## Understanding LEGO **Themes**  VS LEGO **Sets**

###***LEGO SETS***
LEGo *set* is  particular box of LEGO. Therefore a particular theme has a lot of different *sets*.

<img src="https://i.imgur.com/whB1olq.png">

```python
sets = pd.read_csv("sets.csv")
sets.head()
```

```python
sets.sort_values("year").head()
```

```python
sets.sort_values("num_parts", ascending=False).head()
```

```python
sets_by_year = sets.groupby("year").count()
sets_by_year["set_num"].tail()
```

```python
plt.plot(sets_by_year.index[:-2], sets_by_year.set_num[:-2], color = "green")
plt.xlabel('Years')
plt.ylabel('Number of pieces')
plt.show()
```

```python
themes_by_year = sets.groupby('year').agg({'theme_id': pd.Series.nunique})
themes_by_year.rename(columns = {'theme_id': 'number'}, inplace=True)
themes_by_year.head()
```

```python
plt.plot(themes_by_year.index[:-2], themes_by_year.number[:-2], color = ('blue'))
plt.xlabel('Years')
plt.ylabel('Number of pieces')
plt.show()
```

```python
ax1 = plt.gca()
ax2 = plt.twinx()
ax1.plot(sets_by_year.index[:-2], sets_by_year.set_num[:-2], color = "purple")
ax1.set_xlabel('Years')
ax1.set_ylabel('Number of pieces', color = 'green')
ax2.plot(themes_by_year.index[:-2], themes_by_year.number[:-2], color = ('blue'))
ax2.set_ylabel('Number of themes per year', color = 'brown')
```

### Merging (i.e., Combining) DataFrames based on a Key


```python
set_theme_count = pd.DataFrame({'id': set_theme_count.index, 'set_count':set_theme_count.values})
set_theme_count.head()
```

```python
merged = pd.merge(set_theme_count, themes, on='id')
merged[:3]
```

```python
plt.figure(figsize=(14,8))
plt.xticks(fontsize=14, rotation=45)
plt.yticks(fontsize=14)
plt.ylabel('Number of Sets', fontsize=14)
plt.xlabel('Theme Name', fontsize=14)
plt.bar(merged.name[:10], merged.set_count[:10],color= 'red')
plt.show()
```
